```python
# Funcion de recodificación de la variable "Pais de Origen"
def Red(r):
    if r=='Europe':
      r=1
    elif r=='Japan':
      r=2
    elif r=='USA':
      r=3
    else:
      r='No data'
    return r

df1['Pais_Origen']=df1['Pais_Origen'].apply(Red)
df1

```

## Ejercicio #02

```python
rows = len(df1.axes[0]) 
cols = len(df1.axes[1]) 
rows
cols
```

```python
#i = 0
#for i in range(cols):
#   if i == 0:
        
filtered_df = df1.loc[(df1['Anyo']>='1/01/1970')&(df1['Anyo']<'31/12/1975')]

filtered_df.loc[:,(filtered_df!='NaN').all()]
#display(filtered_df)

df2=filtered_df[filtered_df.Millas_Por_Galon.notnull()]
df3=df2[df2.Caballo_De_Fuerza.notnull()]
df3
```

```python
#Item 01
df3['C1']=np.log10(df3['Millas_Por_Galon']/df3['Caballo_De_Fuerza'])

#Item 02
df3['C2']=np.sqrt(df3['Millas_Por_Galon'])*np.exp(df3['Caballo_De_Fuerza'])/200
df3
```

```python
#Item 03
def DEL(a):
  if df3['Millas_Por_Galon'][a]>df3['Caballo_De_Fuerza'][a]:
          i=5
  else:
          i=3
  return i

recorre=range(0,len(df3['Millas_Por_Galon']))
C3=[]

for t in recorre:
  a=DEL(t)
  C3.append(a)

df3['C3']=C3
df3
```

```python
#Item 04
df3['C4']=1/np.log(df3['Millas_Por_Galon']/df3['Caballo_De_Fuerza'])*100

#Item 05
df3['C5']=df3['Caballo_De_Fuerza']**2/df3['Millas_Por_Galon']
df3
```

## Ejercicio #03

```python


Media=round(float(df4.mean()),2)
DesvSt=round(float(df4.std()),2)

LimSup=Media+DesvSt
LimInf=Media-DesvSt

def Var_Cualitativas(i):
  if i>LimSup:
    a='Alto'
  elif i<LimSup and i>LimInf:
    a='Media'
  elif i<LimInf:
    a='Bajo'
  return a

func=range(0,len(df3['Millas_Por_Galon']))

VarGen=[]

for i in func:
  a=Var_Cualitativas(df3.iloc[i,0])
  VarGen.append(a)

df3['Var_Cualitativas']=VarGen
df3
```

## Ejercicio #04

```python
df4=df3[['Millas_Por_Galon','Var_Cualitativas']]
df4

def Grafico(dat):
  sns.set_style('whitegrid')
  fig, axes=plt.subplots(1,2,figsize=(15,5))
  fig.suptitle('Gráficos')
  sns.boxplot(x='Var_Cualitativas',y='Millas_Por_Galon', data=dat, ax=axes[0])
  sns.histplot(data=dat, kde=True,ax=axes[1])

Grafico(df4)
```
